## Expressões regulares
- É uma ferramenta poderosa para buscar e manipular padrões de texto.
- O módulo `re` permite trabalhar com expressões regulares.
- https://docs.python.org/3/library/re.html


```python
### Importando o Módulo re

import re
```

```python
### Encontrando Padrões
"""
A função re.search() verifica se o padrão aparece
em uma string e retorna o primeiro match encontrado.
"""

texto = "A linguagem Python é muito poderosa e possuí vários recursos interessantes."
match = re.search('Python2', texto)

if match:
  print(f'Palavra encontrada: {match.group()}')
else:
  print('Palavra não encontrada!')
```

### Padrões com Caracteres Especiais

#### 1. Metacaracteres Básicos

- **`.`**: Corresponde a qualquer caractere, exceto uma nova linha.
  - Exemplo: `a.b` encontra "a_b", "a*b", "a b", mas não "ab".

- **`^`**: Indica o início da string.
  - Exemplo: `^Olá` corresponde a qualquer string que começa com "Olá".

- **`$`**: Indica o final da string.
  - Exemplo: `mundo$` corresponde a qualquer string que termina com "mundo".

- **`[]`**: Define uma classe de caracteres, ou seja, um conjunto de caracteres permitidos.
  - Exemplo: `[aeiou]` encontra qualquer vogal.

- **`|`**: Operador "ou".
  - Exemplo: `gato|cachorro` encontra "gato" ou "cachorro".

- **`()`**: Agrupa uma expressão, permitindo criar grupos que podem ser referenciados ou capturados.
  - Exemplo: `(abc)+` encontra uma ou mais ocorrências de "abc".

- **`\`**: Caractere de escape, usado para tratar metacaracteres como caracteres literais.
  - Exemplo: `\.` encontra um ponto literal.

#### 2. Classes de Caracteres Especiais

- **`\d`**: Qualquer dígito (equivalente a `[0-9]`).
- **`\D`**: Qualquer caractere que *não* é um dígito.
- **`\w`**: Qualquer caractere alfanumérico (letras, números e underscore).
- **`\W`**: Qualquer caractere que *não* é alfanumérico.
- **`\s`**: Qualquer espaço em branco (espaço, tab, nova linha).
- **`\S`**: Qualquer caractere que *não* é espaço em branco.

#### 3. Quantificadores

Os quantificadores especificam o número de ocorrências do caractere ou grupo que o precede.

- **`*`**: Zero ou mais ocorrências.
  - Exemplo: `a*` encontra "a", "aa", "aaa" ou uma string vazia.
  
- **`+`**: Uma ou mais ocorrências.
  - Exemplo: `a+` encontra "a", "aa", "aaa" (mas não uma string vazia).

- **`?`**: Zero ou uma ocorrência.
  - Exemplo: `a?` encontra "a" ou uma string vazia.

- **`{n}`**: Exatamente `n` ocorrências.
  - Exemplo: `a{3}` encontra "aaa".

- **`{n,}`**: `n` ou mais ocorrências.
  - Exemplo: `a{2,}` encontra "aa", "aaa", etc.

- **`{n,m}`**: Entre `n` e `m` ocorrências.
  - Exemplo: `a{2,4}` encontra "aa", "aaa" ou "aaaa".



```python
# Exemplo
texto = "A linguagem Python está na versão 3.133213132."
match = re.search(r'[0-9].\d{2}', texto)
```

```python
match.group()
```

```python
### Encontrando Múltiplos Resultados

"""
Use re.findall() para retornar todos os matches.
"""

texto = "O carro vermelho possui a placa ABV-1212 e o carro branco possui placa XYZ-9999"
retorno = re.findall(r'\b[A-Z]{3}-\d{4}\b', texto)

print(retorno)
```

```python
### Substituindo Padrões

"""
Use re.sub() para substituir um padrão por outro valor.
"""

texto = "Os telefones do usuário X são: 9999-9999 e 3333-4444."

novo_texto = re.sub(r'[0-9]{2}-[0-9]{4}','####-####', texto)

print(novo_texto)

# Exemplo mais complexo:
# ??99-99?? e ??33-44?? >> 99??-??99 e 33??-??44
# Saída: **??-??**
# novo_texto = re.sub(r'\d{2}-[0-9]{2}','**-**', texto)
```

```python
### Dividindo Strings com re.split()
texto = 'Londrina;Futebol;Café'
palavras = re.split(r';', texto)
print(palavras)

```
